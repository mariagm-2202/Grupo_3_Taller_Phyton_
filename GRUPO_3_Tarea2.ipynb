III. Con ambos dataframes, realiza un inner join. Guarda el resultado en un nuevo dataframe

V. Realizar un gráfico de barras donde se vea la cantidad de habitantes en cada ciudad
image.png

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (8,4))
plt.bar(cuadro_1['Ciudad'],cuadro_1['Habitantes'], color = 'pink')

plt.title('Número de Habitantes por Ciudad en Peru')

plt.xlabel('Ciudad')
plt.ylabel('Habitantes')

plt.show()
```

VI. Realizar un gráfico de dispersión entre la altura y el número de habitantes en las
ciudades del dataframe.

image.png

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (8,4))
plt.scatter(cuadro_1['Altitud_m'],cuadro_1['Habitantes'], color = 'pink')

plt.title('Relacion entre Altitud y Numero de habitantes de las ciudades en Peru')

plt.xlabel('Altitud')
plt.ylabel('Habitantes')

plt.grid(True)
plt.show()
```
