V. Utilizando ùíîùíÑùíäùíëùíö. ùíîùíïùíÇùíïùíî, realiza un an√°lisis de una distribuci√≥n normal con una media de
0 y desviaci√≥n est√°ndar de 1. Genera un conjunto de datos aleatorios con 1000
muestras a partir de esta distribuci√≥n.

```python
# V. An√°lisis de distribuci√≥n normal (media=0, desviaci√≥n est√°ndar=1)
mean = 0
std_dev = 1
data_normal = np.random.normal(mean, std_dev, 1000)
```

VI. Calcula y muestra la media, la desviaci√≥n est√°ndar y la mediana de los datos
generados

```python
# VI. C√°lculo de estad√≠sticos de los datos generados
mean_data = np.mean(data_normal)
std_dev_data = np.std(data_normal)
median_data = np.median(data_normal)

print(f"Media de los datos generados: {mean_data}")
print(f"Desviaci√≥n est√°ndar de los datos generados: {std_dev_data}")
print(f"Mediana de los datos generados: {median_data}")
```

VII. Grafica el histograma de los datos y superp√≥n la curva de la distribuci√≥n normal
te√≥rica.

```python
# VII. Gr√°fico del histograma y la curva de distribuci√≥n normal te√≥rica
plt.figure(figsize=(10, 6))
plt.hist(data_normal, bins=30, density=True, alpha=0.6, color='g', label='Datos generados')

xmin, xmax = plt.xlim()
x_norm = np.linspace(xmin, xmax, 100)
p_norm = norm.pdf(x_norm, mean, std_dev)

plt.plot(x_norm, p_norm, 'k', linewidth=2, label='Distribuci√≥n Normal Te√≥rica')
plt.title('Histograma y Curva de Distribuci√≥n Normal')
plt.xlabel('Valor')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.show()
```
